# Analysing Statistically

```python
#This graph show the distribution of Item_Store_Sales
sns.distplot(bundas_training['Item_Store_Sales'],bins = 10)
```

```python
#Finding the skewness and kurtosis of the item_store_sales
print ("Skew is:", bundas_training.Item_Store_Sales.skew())
print("Kurtosis: %f" % bundas_training.Item_Store_Sales.kurt())

#This results show that the target variable which is Item_Store_Sales is skewed to the higher sales which is the right. 
```

```python
#Selecting the column attributes with numeric features
numeric_features = bundas_training.select_dtypes(include=[np.number])
numeric_features.dtypes
```

```python
#Finding correlation of the numeric columns selected above
corr =numeric_features.corr()
corr
```

```python
#Outputing the correlation from the lowest value to the highest
print(corr["Item_Store_Sales"].sort_values(ascending=True))
```

```python
#The correlation matrix showing the columns and how they relate
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corr, vmax=.8, square=True);

#from this matrix, Visibility has the lowest correlation with Item_Store_Sales
```

```python
#This graph shows the relationship between weight and item store sales. Weight is lowly correlated to item store sales
plt.figure(figsize=(12,7))
plt.xlabel("Weight")
plt.ylabel("Item_Store_Sales")
plt.title("Analyzing Weight and Item_Store_Sales")
plt.plot(bundas_training.Weight, bundas_training["Item_Store_Sales"],'.', color ='gold', alpha = 0.7)
```

```python
#This graph shows the relationship between visibility and item store sales. Weight is lowly correlated to item store sales
plt.figure(figsize=(12,7))
plt.xlabel("Visibility")
plt.ylabel("Item_Store_Sales")
plt.title("Analysis of Item Visibility and Item Store Sales")
plt.plot(bundas_training.Visibility, bundas_training["Item_Store_Sales"],'.', color ='gold', alpha = 0.7)

#this describes the location of the item in the store and how it affects the items bought i.e the ones at the entrance attract customer attention.
```

```python
#This graph shows how the sales were affected in the years the stores were first opened
Store_Establishment_Year_pivot = \
bundas_training.pivot_table(index='Store_Establishment_Year', values="Item_Store_Sales", aggfunc=np.median)
Store_Establishment_Year_pivot.plot(kind='bar', color='gold',figsize=(10,5))
plt.xlabel("Store_Establishment_Year")
plt.ylabel("Sqrt Item_Store_Sales")
plt.title("Impact of Store_Establishment_Year on Item_Store_Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
#Shows the max price of an item against the category
pivot_table_category_prices = \
pivot_table_category_prices = bundas_training.pivot_table(index="Category",values="Max_Price",aggfunc=[np.sum,np.mean])
pivot_table_category_prices["sum"].plot(kind="bar", color ='gold', legend="Sum", figsize=(10,5))
plt.xlabel("Category")
plt.ylabel("Max_Price")
plt.title("Impact of Category on Max Price")
plt.xticks(rotation=90)
plt.show()


```

```python
#shows the impact of Category on Item Store sales using median
pivot_table_category_sales = pd.pivot_table(bundas_training,index=["Category"],values=["Item_Store_Sales"],aggfunc=[np.median])
pivot_table_category_sales["median"].plot(kind="bar", color ='gold', legend="Median", figsize=(10,5))
plt.xlabel("Category")
plt.ylabel("Item_Store_Sales")
plt.title("Impact of Category on Item Store sales using median")
plt.xticks(rotation=90)
plt.show()
```

```python
#shows the impact of Category on Visibility using median
pivot_table_category_visibility = pd.pivot_table(bundas_training,index=["Category"],values=["Visibility"],aggfunc=[np.median])
pivot_table_category_visibility.plot(kind="bar", color ='gold', legend="Median", figsize=(10,5))
plt.xlabel("Category")
plt.ylabel("Visibility")
plt.title("Impact of Category on Visibility using median")
plt.xticks(rotation=90)
plt.show()
```

```python
#shows the impact of Fat Content on Item Store sales using median
pivot_table_fatcontent_sales = pd.pivot_table(bundas_training,index=["FatContent"],values=["Item_Store_Sales"],aggfunc=[np.median])
pivot_table_fatcontent_sales.plot(kind="bar", color ='gold', legend="Median", figsize=(10,5))
plt.xlabel("FatContent")
plt.ylabel("Item_Store_Sales")
plt.title("Impact of Fat Content on Item Store sales using median")
plt.xticks(rotation=0)
plt.show()
```

```python
#shows the StoreIDs and Item Store sales using median meaning, one can tell the sales of each store
pivot_table_storeid_sales = pd.pivot_table(bundas_training,index=["Store_ID"],values=["Item_Store_Sales"],aggfunc=[np.median])
pivot_table_storeid_sales["median"].plot(kind="bar", color ='gold', legend="Median", figsize=(10,5))
plt.xlabel("Store_ID")
plt.ylabel("Item_Store_Sales")
plt.title("Store Identifiers and Item Store Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
#shows the impact of Store size on Item Store sales using median
pivot_table_storesize_sales = pd.pivot_table(bundas_training,index=["Store_Size"],values=["Item_Store_Sales"],aggfunc=[np.median])
pivot_table_storesize_sales["median"].plot(kind="bar", color ='gold', legend="Median", figsize=(10,5))
plt.xlabel("Store_Size")
plt.ylabel("Item_Store_Sales")
plt.title("Impact of Store Size on Item Store sales using median")
plt.xticks(rotation=90)
plt.show()
```

```python
#shows the impact of store type on Item Store sales using median
pivot_table_locationtype_sales = pd.pivot_table(bundas_training,index=["Store_Location_Type"],values=["Item_Store_Sales"],aggfunc=[np.median])
pivot_table_locationtype_sales["median"].plot(kind="bar", color ='gold', legend="Median", figsize=(10,5))
plt.xlabel("Store_Location_Type")
plt.ylabel("Item_Store_Sales")
plt.title("Impact of Store Location Type on Item Store sales using median")
plt.xticks(rotation=90)
plt.show()
```

```python
#focuses on the maximum and minimum - store_type 
pivot_table_itemstore_sales = pd.pivot_table(bundas_training,index=["Store_Type"],values=["Item_Store_Sales"],aggfunc=[np.max,np.min])
print(pivot_table_itemstore_sales.head(10))


# plt.grid()
# plt.plot(pivot_table_item_sales["sum"], 'bs')

pivot_table_itemstore_sales["amax"].plot(kind="bar",legend="Max", color = "gold" , figsize=(10,5)).grid()
```
