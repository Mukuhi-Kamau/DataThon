## 4. Data Pre- Processing

### handling null values

```python
data = pd.concat([bundas_training,bundas_testing], ignore_index = True)
```

```python
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
# Imputation of missing weight values using mean
#pivot table showing weight values based on the item ID
data.pivot_table(values='Weight', index='Item_ID')
```

```python
item_avg_weight = data.pivot_table(values='Weight', index='Item_ID')
```

```python
#function to fill missing values based on the mean of the weight of items with the same item identifier
def impute_weight(cols):
    Weight = cols[0]
    Identifier = cols[1]
    
    if pd.isnull(Weight):
        return item_avg_weight['Weight'][item_avg_weight.index == Identifier]
    else:
        return Weight
print ('Orignal number of null values: %d'%sum(data['Weight'].isnull()))
data['Weight'] = data[['Weight','Item_ID']].apply(impute_weight,axis=1).astype(float)
print ('Final number of null values: %d'%sum(data['Weight'].isnull()))
```

```python
data.isnull().sum()
```

```python
# imputation of store size missing values based on the mode per store type
#Determing the mode for each
store_size_mode = data.pivot_table(values='Store_Size', columns='Store_Type',aggfunc=lambda y:y.mode())
store_size_mode
```

```python
def impute_size_mode(columns):
    Size = columns[0]
    Type = columns[1]
    if pd.isnull(Size):
        return store_size_mode.loc['Store_Size'][store_size_mode.columns == Type][0]
    else:
        return Size
print ('Orignal missing: %d'%sum(data['Store_Size'].isnull()))
data['Store_Size'] = data[['Store_Size','Store_Type']].apply(impute_size_mode,axis=1)
print ('Final missing: %d'%sum(data['Store_Size'].isnull()))
```
